Đề bài: Hàm truyền ở lớp ẩn là tuyến tính, lớp ra là sigmoid

### Truyền thuận 

```python
import torch
import torch.nn as nn
sigmoid = nn.Sigmoid()

input_layer = torch.tensor([1., 0., 1.])

bias_1 = torch.tensor([[-0.4, 0.2]])
weight_1 = torch.tensor([[0.2, -0.3], [0.4, 0.1], [-0.5, 0.2]])
hidden_layer = torch.matmul(input_layer, weight_1)
hidden_layer = torch.add(hidden_layer, bias_1) 

bias_2 = torch.tensor([0.1])
weight_2 = torch.tensor([-0.3, -0.2])
output_layer = torch.matmul(hidden_layer, weight_2)
output_layer = sigmoid(torch.add(output_layer, bias_2))

print(hidden_layer)
print(output_layer)
```
