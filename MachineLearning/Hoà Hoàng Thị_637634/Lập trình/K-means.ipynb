```python
from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from sklearn.cluster import KMeans
from matplotlib import pyplot
# Cài đặt dataset
X, _ = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=2, random_state=None)
# cài đặt model
model = KMeans(n_clusters=2)
# fit model
model.fit(X)
# gán một cụm cho mỗi mẫu
yhat = model.predict(X)
# truy xuất các cụm duy nhất
clusters = unique(yhat)
# tạo biểu đồ phân tán cho các mẫu từ mỗi cụm
for cluster in clusters:
	# lấy chỉ mục hàng cho các mẫu với cụm này
	row_ix = where(yhat == cluster)
	# tạo scatter của những mẫu này
	pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
# show the plot
pyplot.show()
```
