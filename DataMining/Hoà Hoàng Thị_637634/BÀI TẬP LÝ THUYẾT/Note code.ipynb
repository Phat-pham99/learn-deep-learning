### Linear regression

```python
import matplotlib.pyplot as plt

Wheat_production = [30,	28,	32,	25,	25,	25,	22,	24,	35,	40]
Flour_price= [25,	30,	27,	40,	42,	40,	50,	45,	30,	25]

plt.scatter(Wheat_production, Flour_price)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats

Wheat_production = [30,	28,	32,	25,	25,	25,	22,	24,	35,	40]
Flour_price= [25,	30,	27,	40,	42,	40,	50,	45,	30,	25]

slope, intercept, r, p, std_err = stats.linregress(Wheat_production, Flour_price)

def myfunc(Wheat_production):
  return slope * Wheat_production + intercept

mymodel = list(map(myfunc, Wheat_production))

plt.scatter(Wheat_production, Flour_price)
plt.plot(Wheat_production, mymodel)
plt.show()
```
