SVM kernel



```python
# Linear Kernel
K(x, xi) = sum(x * xi)

# Polynomial Kernel
K(x,xi) = 1 + sum(x * xi)^d # d là bậc đa thức

# Radial Basis Function Kernel
K(x,xi) = exp(-gamma * sum((x – xi^2)) # gamma nằm trong khoảng 0->1 thường 0.1
```

Classifier Building in Scikit-learn

###### Load data

```python
# Import scikit-learn dataset 
from sklearn import datasets

# Load dataset
cancer = datasets.load_breast_cancer()
```

###### Khám phá dữ liệu

```python
# in tên 13 features
print("Features: ", cancer.feature_names)

# in nhãn của ung thư là ác tính hay lành tính 
print("Labels: ", cancer.target_names)
```

```python
cancer.data.shape
```

```python
# in ra 5 mẫu
print(cancer.data[0:5])
```

```python
# in labels (0:malignant, 1:benign)
print(cancer.target)
```

###### Splitting Data

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3,random_state=109) # 70% training and 30% test
```

###### Tạo model

```python
# import svm model
from sklearn import svm

# Tạo model phân lớp svm 
clf = svm.SVC(kernel='linear') # Linear Kernel

# train model sử dụng training sets
clf.fit(X_train, y_train)

# dự đoán kết quả trả về cho dataset
y_pred = clf.predict(X_test)
```

###### Đánh giá mô hình

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```
