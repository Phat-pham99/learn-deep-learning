### VD1

```python
import matplotlib.pyplot as plt
from scipy import stats

Wheat_production = [30,	28,	32,	25,	25,	25,	22,	24,	35,	40]
Flour_price= [25,	30,	27,	40,	42,	40,	50,	45,	30,	25]

slope, intercept, r, p, std_err = stats.linregress(Wheat_production, Flour_price)

def myfunc(Wheat_production):
  return slope * Wheat_production + intercept

mymodel = list(map(myfunc, Wheat_production))

plt.scatter(Wheat_production, Flour_price)
plt.plot(Wheat_production, mymodel)
plt.show()
```

```python
print(slope) #B1
print(intercept) #B0
```

### Linear regression algorithm

```python
import numpy as np
import matplotlib.pyplot as plt

def estimate_coef(x, y):
  # Số điểm
  n = np.size(x)

  # Trung bình của vector x và y
  m_x = np.mean(x)
  m_y = np.mean(y)

  # Tính độ lệch chéo và độ lệch của x
  SS_xy = np.sum(y*x) - n*m_y*m_x
  SS_xx = np.sum(x*x) - n*m_x*m_x

  # Tính hệ số tương quan hồi quy
  b_1 = SS_xy / SS_xx
  b_0 = m_y - b_1*m_x

  return (b_0, b_1)

def plot_regression_line(x, y, b):
  # Biểu đồ điểm như biểu đồ phân tán
  plt.scatter(x, y, color = "m", marker = "o", s = 30)

  # Vector trả về dự đoán
  y_pred = b[0] + b[1]*x

  # Vẽ đường tuyến tính
  plt.plot(x, y_pred, color = "g")

  # Đánh nhãn
  plt.xlabel('x')
  plt.ylabel('y')

  # Gọi hàm vẽ
  plt.show()

def main():
  # Data
  x = np.array([30,	28,	32,	25,	25,	25,	22,	24,	35,	40])
  y = np.array([25,	30,	27,	40,	42,	40,	50,	45,	30,	25])

  # Hệ số tương quan
  b = estimate_coef(x, y)
  print("Hệ số tương quan hồi quy là: \nb_0 = {}  \
          \nb_1 = {}".format(b[0], b[1]))
  
  # Vẽ đường tuyến tính
  plot_regression_line(x, y, b)
if __name__ =="__main__":
  main()
```
